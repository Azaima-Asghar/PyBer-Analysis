```python
# Import dependencies.

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a line chart using MATLAB method with plt.plot()

plt.plot(x_axis,y_axis)
plt.xlabel("Month")
plt.ylabel("Total fare ($)")
plt.ylim(0, 45)
plt.plot(x_axis, y_axis, label='Boston')
plt.title("Fare by month")
plt.legend()
plt.plot(x_axis,y_axis,color = "red", marker ="*", linewidth = 2, label="Boston")
plt.show()
```

```python
# Create a line chart --- SKILL DRILL Object oriented.

fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
ax.set_xlabel("Month")
ax.set_ylabel("Total fare ($)")
ax.set_ylim(0, 45)
ax.set_title("PyBer Fare by month")
plt.plot(x_axis,y_axis,color = "green", marker ="d", linewidth = 2, label="Boston")
ax.grid()
ax.legend()
plt.show()
```

```python
# Create a line chart using Object-oriented interface with ax.plt()

fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create a line chart using Object-oriented interface.
# To change figure-level attributes (such as axis labels, a title, or a legend) or 
# save the figure as an image, use fig = plt.figure().

fig = plt.g()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# To create a quick plot with object_oriented interface. 

ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# Create horizontal bar chart -- SKILL DRILL.

plt.barh(x_axis,y_axis, color = "magenta", label = "Boston")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
plt.gca().invert_yaxis()
```

```python
# create a scatter plot.

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
# y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# plt.plot(x_axis,y_axis,"o")

plt.scatter(y_axis,x_axis,s=y_axis)
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0,45)
```

```python
y_axis_bigger = []

for y in y_axis:
    y_axis_bigger.append(y*3)
```

```python
# list comprehension.

plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
plt.show()
```
